```python
import pandas as pd

raw_dataset_path = "/content/Dataset for Data Analytics.xlsx"
cleaned_output_path = "Cleaned_Data_Analytics_Dataset.xlsx"

print(" Loading dataset into memory...")
try:
    df = pd.read_excel(raw_dataset_path)
    print(f" Successfully loaded {len(df)} raw rows.")
except FileNotFoundError:
    print(f" Error: Could not find the file at {raw_dataset_path}. Check your path!")
    exit()

print("\n🧹 Sanitizing and prepping data rules...")

df["CouponCode"] = df["CouponCode"].fillna("NO_COUPON")

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
corrupted_dates_count = df["Date"].isna().sum()

initial_row_count = len(df)
df = df.drop_duplicates(subset=["OrderID"], keep="first")
duplicate_records_removed = initial_row_count - len(df)

print(f"   -> Dropped {duplicate_records_removed} duplicate orders.")
print(f"   -> Flagged {corrupted_dates_count} invalid date strings.")


print("\n[A] UNDERSTANDING THE CENTER OF GRAVITY (Five-Number Summary)")

numerical_summary = df.describe()
print(numerical_summary)

if "Price" in df.columns:
    average_sale_price = df["Price"].mean()
    midpoint_sale_price = df["Price"].median()

    print(f"\n💡 Mathematical Average (Mean) Price : ${average_sale_price:.2f}")
    print(f"💡 Exact Midpoint (Median) Price     : ${midpoint_sale_price:.2f}")
    print("   (Note: If the Mean is much higher than the Median, you have positive skew!)")

if "Category" in df.columns:
    print("\n[B] TRANSACTION VOLUME BY CATEGORY")
    category_trends = df["Category"].value_counts()
    for category, count in category_trends.items():
        print(f"   • {category:<15} : {count} transactions")

if "Price" in df.columns:
    print("\n[C] HUNTING FOR HIGHEST-VALUE TRANSACTION ANOMALIES (Outliers)")
    top_outliers = df.sort_values(by="Price", ascending=False).head(5)

    for idx, row in top_outliers.iterrows():
        print(f"   💥 Order ID: {row['OrderID']} | Price: ${row['Price']:.2f}")


df.to_excel(cleaned_output_path, index=False)
print(f"🎉 Analysis ready! Cleaned data saved to: '{cleaned_output_path}'")
```
